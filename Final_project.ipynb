```python
import speech_recognition as speech_recog
from gingerit.gingerit import GingerIt
import nltk
from nltk import word_tokenize
import string
import pandas as pd
import openpyxl

#音檔轉換文字
recog = speech_recog.Recognizer()

results=[]
student=[]
for i in range(2):
    file_path = '/Users/huangtsaiyuan/Desktop/School/111-2/物聯網/{}.wav'.format(i)
    sample_audio = speech_recog.AudioFile(file_path)
    with sample_audio as audio_file:
        recog.adjust_for_ambient_noise(audio_file)
        audio_content = recog.record(audio_file)
        result=recog.recognize_google(audio_content)
        results.append(result)
        student.append(i)

#Grammar checker
Grammer = GingerIt()
num=0
er=[]
cor=[]
for t in results:
    Correction = Grammer.parse(t)
    cor.append(Correction)
    er.append(len(Correction['corrections']))
    i+=1


#TTR
TTR=[]
#assign the string of common punctuation symbols to a variable and turn it into a list
punctuations = list(string.punctuation)
#create new variable that applies the word_tokenize function to our sentence.
for t in results:
    sentence_tokens = word_tokenize(t)
    sentence_tokens_clean = [word for word in sentence_tokens if word not in punctuations]
    TTR.append(round(len(list(set(sentence_tokens_clean)))/len(sentence_tokens_clean), 3))


df = pd.DataFrame()
df['Student'] = student
df["Grammatical Error"] = er
df["TTR"] = TTR
df["Correction"] = cor
df.to_excel("output.xlsx",index=False)
```
